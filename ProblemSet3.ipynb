Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
def fix_century(x):
  year = x.year - 100 if x.year > 1989 else x.year
  return datetime.date(year, x.month, x.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century)

data.head()
```

# <b>Question 6</b>

Create a line plot showing the number of marriages and divorces per capita in theU.S. between 1867 and 2014. Label both lines and show the legend.

Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
us_marriage_divorce_data = pd.read_csv("C:/Users/parth/OneDrive/Desktop/BDAT/Sem 1/Data Programming/Problem Set 3/us-marriages-divorces-1867-2014.csv")

# Extract relevant columns
years = us_marriage_divorce_data['Year'].values
marriages_per_capita = us_marriage_divorce_data['Marriages_per_1000'].values
divorces_per_capita = us_marriage_divorce_data['Divorces_per_1000'].values

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(years, marriages_per_capita, label='Marriages per Capita', color='blue')
plt.plot(years, divorces_per_capita, label='Divorces per Capita', color='red')

# Labeling
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.xlabel('Year')
plt.ylabel('Per 1000 People')
plt.legend()

plt.show()
```

# <b>Question 7</b>

Create a vertical bar chart comparing the number of marriages and divorces percapita in the U.S. between 1900, 1950, and 2000.

Don't forget to label your axes

```python
selected_years = [1900, 1950, 2000]
selected_data = us_marriage_divorce_data[us_marriage_divorce_data['Year'].isin(selected_years)]

# Create a vertical bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.35
bar_positions = selected_data.index

plt.bar(bar_positions - bar_width/2, selected_data['Marriages_per_1000'], bar_width, label='Marriages per Capita', color='blue')
plt.bar(bar_positions + bar_width/2, selected_data['Divorces_per_1000'], bar_width, label='Divorces per Capita', color='red')

# Labeling
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
```

# <b>Question 9</b> 

Create a pie chart showing the fraction of all Roman Emperors that were assassinated.

Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the file
file_path = r'C:/Users/parth/OneDrive/Desktop/BDAT/Sem 1/Data Programming/Problem Set 3/roman-emperor-reigns.csv'
data = pd.read_csv(file_path)

# Get unique categories of the cause of death
death_categories = data['Cause_of_Death'].unique()

# Count the number of emperors for each category
death_counts = data['Cause_of_Death'].value_counts()

# Data for the pie chart
labels = death_counts.index
sizes = death_counts.values

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.tab10.colors)

# Labeling
plt.title('Cause of Death of Roman Emperors')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is a circle

# Show the pie chart
plt.show()
```

# <b>Question 10</b>

Create a scatter plot showing the relationship between the total revenue earned byarcades and the number of Computer Science PhDs awarded in the U.S. between2000 and 2009.

Don't forget to label your axes!

Color each dot according to its year.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:/Users/parth/OneDrive/Desktop/BDAT/Sem 1/Data Programming/Problem Set 3/arcade-revenue-vs-cs-doctorates.csv'
data = pd.read_csv(file_path)

# Create a scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    data['Total Arcade Revenue (billions)'],
    data['Computer Science Doctorates Awarded (US)'],
    c=data['Year'],  # Color each dot according to its year
    cmap='viridis',  
    alpha=0.8,
    edgecolors='w',  # Add white edge colors for better visibility
)

# Labeling
plt.title('Relationship between Arcade Revenue and Computer Science PhDs (2000-2009)')
plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Computer Science Doctorates Awarded (US)')
plt.colorbar(scatter, label='Year')

# Show the plot
plt.show()
```
